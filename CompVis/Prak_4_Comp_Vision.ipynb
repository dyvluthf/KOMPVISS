```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# ===============================
# 1. Upload gambar
# ===============================
uploaded = files.upload()
file_names = list(uploaded.keys())

print("File yang diupload:", file_names)

# ===============================
# 2. Cek jumlah file
# ===============================
if len(file_names) == 0:
    print("❌ Tidak ada file diupload!")

elif len(file_names) == 1:
    print("⚠️ Hanya 1 gambar, akan dipakai untuk keduanya")

    image1 = cv2.imread(file_names[0], 0)
    image2 = cv2.imread(file_names[0], 0)

else:
    image1 = cv2.imread(file_names[0], 0)
    image2 = cv2.imread(file_names[1], 0)

# ===============================
# 3. Cek apakah gambar terbaca
# ===============================
if image1 is None or image2 is None:
    print("❌ Gambar tidak terbaca! Cek nama file.")
else:

    # ===============================
    # 4. Threshold
    # ===============================
    thresholds = [50, 100, 150, 200]

    def tampilkan(image, title):
        plt.figure(figsize=(10,6))
        plt.suptitle(title)

        for i, t in enumerate(thresholds):
            _, th = cv2.threshold(image, t, 255, cv2.THRESH_BINARY)
            plt.subplot(2,2,i+1)
            plt.imshow(th, cmap='gray')
            plt.title(f'Threshold {t}')
            plt.axis('off')

        plt.show()

    # ===============================
    # 5. Tampilkan hasil
    # ===============================
    tampilkan(image1, "Gambar 1")
    tampilkan(image2, "Gambar 2")

    # ===============================
    # 6. Info citra
    # ===============================
    print("Shape Gambar 1:", image1.shape)
    print("Dtype Gambar 1:", image1.dtype)

    print("Shape Gambar 2:", image2.shape)
    print("Dtype Gambar 2:", image2.dtype)
```
