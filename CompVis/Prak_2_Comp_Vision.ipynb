```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# ===============================
# 1. Upload gambar
# ===============================
uploaded = files.upload()

# Ambil nama file yang diupload
file_names = list(uploaded.keys())

# Pastikan minimal ada 1 file
if len(file_names) == 0:
    print("Tidak ada file diupload!")
else:
    print("File yang diupload:", file_names)

# ===============================
# 2. Gunakan file
# ===============================
# Jika hanya 1 file → pakai untuk PNG & JPG
image_png = cv2.imread(file_names[0])
image_jpg = cv2.imread(file_names[0])

# ===============================
# 3. Cek apakah gambar terbaca
# ===============================
if image_png is None or image_jpg is None:
    print("Gambar tidak terbaca! Cek nama file.")
else:
    # ===============================
    # 4. Tampilkan citra
    # ===============================
    plt.subplot(1,2,1)
    plt.imshow(cv2.cvtColor(image_png, cv2.COLOR_BGR2RGB))
    plt.title('PNG')
    plt.axis('off')

    plt.subplot(1,2,2)
    plt.imshow(cv2.cvtColor(image_jpg, cv2.COLOR_BGR2RGB))
    plt.title('JPG')
    plt.axis('off')

    plt.show()

    # ===============================
    # 5. Tampilkan informasi
    # ===============================
    print("PNG Shape :", image_png.shape)
    print("PNG Dtype :", image_png.dtype)

    print("JPG Shape :", image_jpg.shape)
    print("JPG Dtype :", image_jpg.dtype)
```
